```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_function(x):
    return 2*x**2 + 3*x + 1

x = np.linspace(-10, 10, 100)

plt.plot(x, quadratic_function(x))

plt.xlabel(' x')
plt.ylabel(' y')
plt.title('  ')

plt.show()
```

```python
def harmonic_function(x):
    return np.sin(x)

plt.plot(x, harmonic_function(x))

plt.xlabel(' x')
plt.ylabel(' y')
plt.title('  ')

plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, quadratic_function(x))
plt.title(' ')

plt.subplot(1, 2, 2)
plt.plot(x, harmonic_function(x))
plt.title(' ')

plt.show()
```

```python
plt.plot(x, quadratic_function(x), label=' ')
plt.plot(x, harmonic_function(x), label=' ')

plt.xlabel(' x')
plt.ylabel(' y')
plt.title('    ')

plt.legend()

plt.show()
```

```python
t = np.linspace(0, 2*np.pi, 100)

plt.plot(np.cos(t), np.sin(t))

plt.xlabel(' x')
plt.ylabel(' y')
plt.title('  ')

plt.show()
```

```python
categories = [' 1', ' 2', ' 3']
values = [15, 30, 45]

plt.bar(categories, values)

plt.xlabel('')
plt.ylabel('')
plt.title(' ')

plt.show()
```

```python
labels = [' 1', ' 2', ' 3']
sizes = [25, 40, 35]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.title(' ')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title(' ')

plt.show()
```
